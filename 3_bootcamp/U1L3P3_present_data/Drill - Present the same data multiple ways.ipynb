```python
iris = sns.load_dataset("iris")
print(iris.head(n=4))
```

Choose one variable and plot that variable four different ways.

```python
x = iris.sepal_length

plt.figure()
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15,10))
sns.distplot(x, ax=ax1)
sns.boxplot(x, orient='v', ax=ax2)
sns.violinplot(x, orient='v', ax=ax3)
sns.kdeplot(x, ax=ax4)
plt.show()
```

Choose two continuous variables, and plot them three different ways.

```python
y = iris.petal_width

plt.figure()
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10,10))
sns.regplot(x, y, ax=ax1)
sns.distplot(x, ax=ax2)
sns.distplot(y, ax=ax2, color='g')
ax2.set(xlabel='width', ylabel='')
sns.violinplot(x, ax=ax3)
sns.violinplot(y, ax=ax3, color='g')
ax3.set(xlabel='width', ylabel='')
plt.show()
```

Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
setosa = iris[iris['species'] == 'setosa']
virginica = iris[iris['species'] == 'virginica']
versicolor = iris[iris['species'] == 'versicolor']

plt.figure()
sns.distplot(setosa.sepal_length)
sns.distplot(virginica.sepal_length)
sns.distplot(versicolor.sepal_length)
plt.show()
```

```python
plt.figure()
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, figsize=(10,10))

sns.barplot(x='species', y='sepal_length', data=iris, ax=ax1)
sns.boxplot(x='species', y='sepal_length', data=iris, ax=ax2)
sns.barplot(x='species', y='sepal_length', data=iris, ax=ax3)
sns.violinplot(x='species', y='sepal_length', data=iris, ax=ax4)
plt.show()
```
