Using your own Jupyter notebook, or a copy of the notebook from the previous assignment, reproduce the pop1 and pop2 populations and samples, using numpy's binomial function. Specifically, create two binomially distributed populations with n equal to 10 and size equal to 10000. The p-value of pop1 should be 0.2 and the p-value of pop2 should be 0.5. Using a sample size of 100, calculate the means and standard deviations of your samples.

```python
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10, 0.2, 10000)

```

```python
def analyze_distribution(size):
    sample1 = np.random.choice(pop1, size, replace=True)
    sample2 = np.random.choice(pop2, size, replace=True)
    for dist in (sample1, sample2):
        print('mean: ', dist.mean())
        print('std: ', dist.std())
        
        plt.hist(dist, color='orange')
        plt.axvline(dist.mean(), color='b', linestyle='solid', linewidth=2)
        plt.axvline(dist.mean() + dist.std(), color='b', linestyle='dashed', linewidth=2)
        plt.axvline(dist.mean() - dist.std(), color='b', linestyle='dashed', linewidth=2)
        plt.show()
    
```

```python
analyze_distribution(100)
```
