```python
import itertools
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
from sklearn import datasets
from sklearn.decomposition import PCA as sklearnPCA
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

### Assigment 1
Plot the Iris Dataset by two attributes that you found to cluster well in Lesson 2, and color by species.

```python
iris = datasets.load_iris()
```

```python
flowers = list(set(iris.target))
fields = [0, 1, 2, 3]
field_combs = []
for field in itertools.combinations(fields, 2):
    field_combs.append(field)
```

```python
attribute_X = field_combs[5][0]
attribute_Y = field_combs[5][1]
rows = []
for flower in flowers:
    row = iris.target==flower
    rows.append(row)

for i, col in zip(range(0,3), ('blue', 'red', 'green')):    
    scatter = plt.scatter(iris.data[rows[i], attribute_X], iris.data[rows[i], attribute_Y], c=col, label=iris.target_names[i])
plt.xlabel(iris.feature_names[field_combs[5][0]])
plt.ylabel(iris.feature_names[field_combs[5][1]])
plt.legend(loc='upper left')
plt.show()
```

### Assigment 2

```python
# Perform PCA on the Iris Dataset.
X_std = StandardScaler().fit_transform(iris.data)
sklearn_pca = sklearnPCA(n_components=2)
Y_sklearn = sklearn_pca.fit_transform(X_std)
```

```python
# Plot the PCA sample.
plt.figure(figsize=(6, 4))
for targ, lab, col in zip((0, 1, 2),
                    ('Iris-setosa', 'Iris-versicolor', 'Iris-virginica'),
                    ('blue', 'red', 'green')):
    plt.scatter(Y_sklearn[iris.target==targ, 0],
                Y_sklearn[iris.target==targ, 1],
                label=lab,
                c=col)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(loc='upper center')
plt.tight_layout()
plt.show()
```

```python
#Retrain the k-nearest neighbors algorithm from Lesson 5 on the decomposed dataset.
neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(Y_sklearn,iris.target)
model_1 = neigh.predict(Y_sklearn)

# Compare the performance on the decomposed dataset to the performance on the raw dataset. Does anything stand out?
neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(iris.data,iris.target)
model_2 = neigh.predict(iris.data)

errors_1 = 0
errors_2 = 0
for real, guess1, guess2 in zip(iris.target, model_1, model_2):
    if real != guess1:
        errors_1 +=1
    if real != guess2:
        errors_2 +=1
print """Number of flowers incorrectly tagged by each model:
model_1: {}
model_2: {}
        """.format(errors_1, errors_2)
```

### Assigment 3

```python
# Perform LDA on the Iris Dataset.
sklearn_lda = LDA(n_components=2)
X_lda_sklearn = sklearn_lda.fit_transform(iris.data, iris.target)

#Plot the LDA sample. How does it compare to the original plot?

for label, name, marker, color in zip(
    range(0,4),('Setosa', 'Versicolor', 'Virginica'), ('^', 's', 'o'),('blue', 'red', 'green')):

    plt.scatter(x=iris.data[:,0][iris.target == label],
                y=iris.data[:,1][iris.target == label] * -1,
                marker=marker,
                color=color,
                alpha=0.5,
                label=name)

plt.xlabel('LD1')
plt.ylabel('LD2')

leg = plt.legend(loc='upper right', fancybox=True)
leg.get_frame().set_alpha(0.5)

# hide axis ticks
plt.tick_params(axis="both", which="both", bottom="off", top="off",  
        labelbottom="on", left="off", right="off", labelleft="on")
  

plt.grid()
plt.tight_layout
plt.show()
```

```python
#Retrain the k-nearest neighbors algorithm from Lesson 4 on the decomposed dataset.
neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(X_lda_sklearn,iris.target)
model_1 = neigh.predict(X_lda_sklearn)

# Compare the performance on the decomposed dataset to the performance on the raw dataset. Does anything stand out?
neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(iris.data,iris.target)
model_2 = neigh.predict(iris.data)

errors_1 = 0
errors_2 = 0
for real, guess1, guess2 in zip(iris.target, model_1, model_2):
    if real != guess1:
        errors_1 +=1
    if real != guess2:
        errors_2 +=1
print """Number of flowers incorrectly tagged by each model:
model_1: {}
model_2: {}
        """.format(errors_1, errors_2)
```
